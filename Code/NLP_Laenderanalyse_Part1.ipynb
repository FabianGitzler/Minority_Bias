## Türkei

```python
df_türkei
```

```python
# Schreibt die Query Expansions in eine eigene Liste
list_queryExpansions = df_türkei["cleaned_data"].to_list()

# Zerlegt die Liste in kleinere Listen um den Rechner nicht zu überfordern
binned_queryExpansions = [list_queryExpansions[i:i + 500] for i in range(0, len(list_queryExpansions), 500)]
```

```python
# Leere Liste in der die Results gespeichert werden
list_results =[]

# Geht durch die Listen in der Liste, bewertet die darin gespeicherten Elemente 
# und speichert das Ergebnis in der leeren Liste
for i in binned_queryExpansions:
    result = model.predict_sentiment(i) 
    list_results.append(result)
```

```python
# Macht aus Listen in Liste eine flache Liste, damit Counter diese lesen kann
list_results = [item for sublist in list_results for item in sublist]

# Counter zählt die Häufigkeit aller positiv, negativ und neutral bewerteten Query Expansions 
result_final = Counter(list_results)

# Wandelt das collections.Counter-Objekt in ein Dictionary, um daraus einen  Dataframe erstellen zu können
dict_result = dict(Counter(result_final))

# Erzeugt aus aus dem Dictionary einen Dataframe
df_results = pd.DataFrame.from_dict([dict_result])
```
