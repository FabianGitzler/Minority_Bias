# CSVs der Länderanalysen zusammenfügen

```python
# Imports
import pandas as pd
```

```python
# Schreibt alle Länderanalysen in eine Liste
li =[]

df1 = pd.read_csv('D:/TH_Koeln/SS21/DIS25_Natural_Language_Processing/Abgabe/NLP_Laenderanalyse_Part1.csv', index_col=None)
li.append(df1)

df2 = pd.read_csv('D:/TH_Koeln/SS21/DIS25_Natural_Language_Processing/Abgabe/NLP_Laenderanalyse_Part2.csv', index_col=None)
li.append(df2)

df3 = pd.read_csv('D:/TH_Koeln/SS21/DIS25_Natural_Language_Processing/Abgabe/NLP_Laenderanalyse_Part3.csv', index_col=None)
li.append(df3)
```

```python
# Fügt den Inhalt der Liste zu einem Datafram zusammen
df = pd.concat(li, axis=0, ignore_index=True)
```

```python
# Speichert den Dataframe in einer CSV
df.to_csv('D:/TH_Koeln/SS21/DIS25_Natural_Language_Processing/Abgabe/NLP_Laenderanalyse_Komplett.csv', index=False)
```
