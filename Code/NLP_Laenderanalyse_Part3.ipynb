## Erstellung der Länder-Dataframes

```python
df_türkei = df[df['queryterm'].str.contains(r'türk', na=False)]
df_polen = df[df['queryterm'].str.contains(r'pol', na=False)]
df_rumänien = df[df['queryterm'].str.contains(r'rumän', na=False)]
df_griechenland = df[df['queryterm'].str.contains(r'griech', na=False)]
df_kroatien = df[df['queryterm'].str.contains(r'kroat', na=False)]
df_bulgarien = df[df['queryterm'].str.contains(r'bulgar', na=False)]
df_afghanistan = df[df['queryterm'].str.contains(r'afghan', na=False)]
df_russland = df[df['queryterm'].str.contains(r'russ', na=False)]
df_serbien = df[df['queryterm'].str.contains(r'serb', na=False)]
df_portugal = df[df['queryterm'].str.contains(r'portugies', na=False)]
df_frankreich = df[df['queryterm'].str.contains(r'franz', na=False)]
df_china = df[df['queryterm'].str.contains(r'chines', na=False)]
df_vietnam = df[df['queryterm'].str.contains(r'vietnames', na=False)]
df_pakistan = df[df['queryterm'].str.contains(r'pakistan', na=False)]
df_irak = df[df['queryterm'].str.contains(r'irak', na=False)]
df_niederlande = df[df['queryterm'].str.contains(r'holländ|niederländ', na=False)]
df_österreich = df[df['queryterm'].str.contains(r'österreich', na=False)]
df_bosnien = df[df['queryterm'].str.contains(r'bosni', na=False)]
df_spanien = df[df['queryterm'].str.contains(r'spani', na=False)]
df_ukraine = df[df['queryterm'].str.contains(r'ukrain', na=False)]
df_syrien = df[df['queryterm'].str.contains(r'syr', na=False)]
df_italien = df[df['queryterm'].str.contains(r'italien', na=False)]
df_amerika = df[df['queryterm'].str.contains(r'amerikan', na=False)]
df_england = df[df['queryterm'].str.contains(r'engländer|englisch', na=False)]
df_schweiz = df[df['queryterm'].str.contains(r'schweizer', na=False)]
df_arabisch = df[df['queryterm'].str.contains(r'arab', na=False)]
df_afrika = df[df['queryterm'].str.contains(r'afrika', na=False)]
df_iran = df[df['queryterm'].str.contains(r'iran', na=False)]
df_indien = df[df['queryterm'].str.contains(r'inder|indisch', na=False)]
df_israel = df[df['queryterm'].str.contains(r'israel', na=False)]
df_deutschland = df[df['queryterm'].str.contains(r'deutsch', na=False)]
```

## Syrien

```python
# Schreibt die Query Expansions in eine eigene Liste
list_queryExpansions = df_syrien["cleaned_data"].to_list()

# Zerlegt die Liste in kleinere Listen um den Rechner nicht zu überfordern
binned_queryExpansions = [list_queryExpansions[i:i + 500] for i in range(0, len(list_queryExpansions), 500)]
```

```python
# Leere Liste in der die Results gespeichert werden
list_results =[]

# Geht durch die Listen in der Liste, bewertet die darin gespeicherten Elemente 
# und speichert das Ergebnis in der leeren Liste
for i in binned_queryExpansions:
    result = model.predict_sentiment(i) 
    list_results.append(result)
```

```python
# Macht aus Listen in Liste eine flache Liste, damit Counter diese lesen kann
list_results = [item for sublist in list_results for item in sublist]

# Counter zählt die Häufigkeit aller positiv, negativ und neutral bewerteten Query Expansions 
result_final = Counter(list_results)

# Wandelt das collections.Counter-Objekt in ein Dictionary, um daraus einen  Dataframe erstellen zu können
dict_result = dict(Counter(result_final))

# Erzeugt aus aus dem Dictionary einen Dataframe
df_results = pd.DataFrame.from_dict([dict_result])
```

```python
df_results
```
