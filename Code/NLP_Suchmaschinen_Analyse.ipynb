# Suchmaschinenanalyse bezogen auf religiöse Gruppen

```python
import pandas as pd
import re
from germansentiment import SentimentModel
from collections import Counter
```

* Die Notwendigen Imports sind Pandas, re für Regex, SentimentModel für Sentimentbestimmung und Counter zum Zählen von Listenelementen

```python
# Zeigt Alle Spalten und Zeilen in voller Länge
pd.set_option('display.max_colwidth', None)
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
# Einlesen der gesäuberten CSV
df = pd.read_csv('D:/TH_Koeln/SS21/DIS25_Natural_Language_Processing/Abgabe/NLP_All_Minorities_Cleaned.csv', index_col=None)
```

```python
df
```

```python
model = SentimentModel() #Aktivierung von germansentiment
```

# Religiöse Gruppen erstellen

```python
df_jüdisch = df[df['queryterm'].str.contains(r'jude|jüdisch', na=False)]
df_islam = df[df['queryterm'].str.contains(r'muslim|islam', na=False)]
df_christ = df[df['queryterm'].str.contains(r'christ|protestant|kathol|evangel', na=False)]
```

* Erstellung von einzelnen Dataframes, um zu analysieren, inwiefern ein Bias in verschiedenen Suchmaschinen vorliegt.

# Zusätzliche Gruppe erstellen

```python
df_flücht = df[df['queryterm'].str.contains(r'migrant|immigrant|flüchtling|illegaler', na=False)]
```

* Zusätzlich werden Queryterms analysiert, die etwas mit dem Thema "Flüchtlinge" zu tun haben
* "df_nachbar" und "df_fern" wurden verworfen

# Analyse
# Dataframe "df_jüdisch"

## Bing

```python
df_jüdisch_bing
```

### Datensatz nach Bing filtern

```python
df_jüdisch_bing = df_jüdisch[df_jüdisch['search_engine'].str.contains(r'bing', na=False)]
```

### Query Expansions in eigene Liste schreiben und in Sub-Listen aufteilen, damit nicht zuviel Arbeitsspeicher bei der Analyse beansprucht wird

```python
list_jüdisch_bing = df_jüdisch_bing["cleaned_data"].to_list()
jüdisch_bing_binned = [list_jüdisch_bing[i:i + 500] for i in range(0, len(list_jüdisch_bing), 500)]
```

### Schleife zur Durchführung der Sentiment-Analyse

```python
list_jüdisch_bing_results =[]

for i in jüdisch_bing_binned:
    result_jüdisch_bing = model.predict_sentiment(i) 
    list_jüdisch_bing_results.append(result_jüdisch_bing)
```

### Ergebnisse in eigene Liste schreiben und die Listenelemente zählen

```python
list_jüdisch_bing_results = [item for sublist in list_jüdisch_bing_results for item in sublist]
result_jüdisch_bing = Counter(list_jüdisch_bing_results)
result_jüdisch_bing
```

## df_jüdisch_bing Verteilung:
* negativ: 227
* positiv: 17
* neutral: 1259

### Das Ergebnis wird in ein neues Dataframe zur Weiterverarbeitung geladen

```python
dict_result = {'neutral': 1259, 'negative':227, 'positive':17, 'searchengine':"bing", 'dataframe':"jüdisch"}
df_results = pd.DataFrame.from_dict([dict_result])
df_results
```
