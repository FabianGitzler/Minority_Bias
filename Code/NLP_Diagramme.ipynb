# Visualisierung der Ergebnisse für Suchmaschinen

```python
import numpy as np
```

## Suchmaschinen

```python
import pandas as pd
import matplotlib.pyplot as plt


# Diagramm für die Suchmaschine "Bing"
# Werte für unterschiedliche Stimmungslagen 
labels = ['negativ', 'neutral', 'positiv']  
# Prozentwerte
sizes = [10.1, 89.58, 0.32] 
# Farbgebung für alle 3 Teile
colors = ['darkred', 'lightgrey', 'seagreen'] 


# Verschiedene Details wie Textgröße und andere verschiedene Elemente
# Daten
plt.pie(sizes, 
        # Beschriftung der Anteile
        labels=labels,
        # Array von Farben
        colors=colors, 
        # Anfangswinkel für Kreisdiagramm
        startangle=140)    

# Diagrammtitel
plt.title('Sentimentverteilung für Bing') 

# legende erzeugen
plt.legend(sizes, 
          # Legendentitel
          title ="Anteil in %", 
          # Ausrichtung
          loc ="center left", 
          bbox_to_anchor =(1.0, 0, 1, 1.5)) 

# Diagramm anzeigen
plt.show() 

# Diagramm für für die Suchmaschine "Google"
labels = ['negativ', 'neutral', 'positiv']
sizes = [5.38, 94.34, 0.28]
colors = ['darkred', 'lightgrey','seagreen']

plt.pie(sizes,             
        labels=labels,     
        colors=colors,      
        startangle=140)     

plt.legend(sizes, 
          title ="Anteil in %", 
          loc ="center left", 
          bbox_to_anchor =(1.0, 0, 1, 1.5)) 

plt.title('Sentimentverteilung für Google')

plt.show()


# Diagramm für für die Suchmaschine"DDG"
labels = ['negativ', 'neutral', 'positiv']
sizes = [19.2, 79.16,1.6]
colors = ['darkred', 'lightgrey','seagreen']

plt.pie(sizes,              
        labels=labels,      
        colors=colors,      
        startangle=140)     

plt.legend(sizes, 
          title ="Anteil in %", 
          loc ="center left", 
          bbox_to_anchor =(1.0, 0, 1, 1.5)) 

 
plt.title('Sentimentverteilung für Duck Duck Go')

plt.show()
```

## Religionen

```python
# -!Kommentare sind für alle folgenden Diagramme gültig!-
import matplotlib.pyplot as plt

# Beschriftung der Balken
labels = ['Google', 'Bing', 'Ddg']
# Werte für die unterschiedlichen Stimmungslagen
positiv = [10,17,87]
neutral = [2531,1259,2113]
negativ = [238,227,540]

# Breite der Säulen
width = 0.50       

fig, ax = plt.subplots()
# Position der Stimmungslagen
ax.bar(labels,negativ, width, label='negativ', bottom=neutral,color='darkred')
ax.bar(labels,neutral, width, label='neutral',color='lightgrey')
# Verwendung von Array um Überlappung zu vermeiden und exakte Angrenzung zu schaffen
ax.bar(labels,positiv, width, label='positiv', bottom=(np.array(negativ)+np.array(neutral)),color='seagreen')

# Beschriftung der Y-Achse
ax.set_ylabel('Anzahl Queryterms')
# Diagrammtitel
ax.set_title('Sentimentverteilung Queryterm "jüdisch" nach Suchmaschinen')
# Passende Legende erzeugen
ax.legend()
# Diagramm anzeigen
plt.show() 


# Diagramm für Islam
positiv = [25,1,118]
neutral = [3786,1338,1465]
negativ = [478,231,1065]

fig, ax = plt.subplots()

ax.bar(labels,negativ, width, label='negativ', bottom=neutral,color='darkred')
ax.bar(labels,neutral, width, label='neutral',color='lightgrey')
ax.bar(labels,positiv, width, label='positiv', bottom=(np.array(negativ)+np.array(neutral)),color='seagreen')

ax.set_ylabel('Anzahl Queryterms')
ax.set_title('Sentimentverteilung Queryterm "Islam" nach Suchmaschinen')
ax.legend()

plt.show()



# Diagramm für Christ
positiv = [26,7,31]
neutral = [9202,3012,5598]
negativ = [133,207,420]

fig, ax = plt.subplots()

ax.bar(labels,negativ, width, label='negativ', bottom=neutral,color='darkred')
ax.bar(labels,neutral, width, label='neutral',color='lightgrey')
ax.bar(labels,positiv, width, label='positiv', bottom=(np.array(negativ)+np.array(neutral)),color='seagreen')


ax.set_ylabel('Anzahl Queryterms')
ax.set_title('Sentimentverteilung Queryterm "Christ" nach Suchmaschinen')
ax.legend()

plt.show()


# Diagramm für flücht
positiv = [2,1,9]
neutral = [6036,1636,2638]
negativ = [306,226,840]

fig, ax = plt.subplots()

ax.bar(labels,negativ, width, label='negativ', bottom=neutral,color='darkred')
ax.bar(labels,neutral, width, label='neutral',color='lightgrey')
ax.bar(labels,positiv, width, label='positiv', bottom=(np.array(negativ)+np.array(neutral)),color='seagreen')

ax.set_ylabel('Anzahl Queryterms')
ax.set_title('Sentimentverteilung Queryterm "Flüchtling" nach Suchmaschinen')
ax.legend()

plt.show()
```
